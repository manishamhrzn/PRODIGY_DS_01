Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from CSV file
population = pd.read_csv('/Users/Manisha/DS_01/world_population.csv')
population
```

```python
population.info()
```

```python
population.describe()
```

```python
df = pd.DataFrame(population)
Continent = df['Continent']
pop = df['2022 Population']
plt.bar(Continent, pop, width=0.25, color="green")
plt.show()
```

```python
plt.hist(Continent, bins=15, rwidth=1000)
```

```python
sns.barplot(x='Continent', y='2022 Population', data=population)
```

```python
sns.histplot(data = population, x = "Continent", kde = True)
```
